```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
data = pd.read_csv("RAG_data.csv")
```

```python
# 1. Соотношение по источникам
source_counts = data["ИсточникФайл"].value_counts()
source_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Распределение вопросов по источникам")
plt.show()
```

```python

# 1. Подсчет распределения по источникам
source_counts = data["ИсточникФайл"].value_counts(normalize=True) * 100

# 2. Форматирование результатов
report = """
РАСПРЕДЕЛЕНИЕ ВОПРОСОВ ПО ИСТОЧНИКАМ
-------------------------------------
"""

for source, percent in source_counts.items():
    report += f"- {source}: {percent:.1f}% вопросов\n"

# 3. Анализ распределения
total_sources = len(source_counts)
dominant_source = source_counts.idxmax()
dominant_percent = source_counts.max()
minor_source = source_counts.idxmin()
minor_percent = source_counts.min()

report += f"""
АНАЛИЗ РАСПРЕДЕЛЕНИЯ
-------------------
1. Всего категорий источников: {total_sources}
2. Наибольшая доля: {dominant_source} ({dominant_percent:.1f}%)
3. Наименьшая доля: {minor_source} ({minor_percent:.1f}%)
4. Соотношение наибольшей/наименьшей доли: {dominant_percent/minor_percent:.1f}:1

КЛЮЧЕВЫЕ НАБЛЮДЕНИЯ:
- {dominant_source} доминирует, составляя {dominant_percent:.1f}% всех вопросов
- {' и '.join(source_counts.index[1:3].tolist())} вместе составляют {source_counts.iloc[1:3].sum():.1f}%
- Распределение показывает {'перекос' if dominant_percent > 50 else 'баланс'} в сторону {dominant_source.split('.')[0]}
"""

# 4. Вывод результатов
print(report)

# 5. Дополнительная статистика (опционально)
stats = {
    'Общее количество вопросов': len(data),
    'Среднее вопросов на источник': len(data)/total_sources,
    'Медианная доля': source_counts.median()
}

print("\nДОПОЛНИТЕЛЬНАЯ СТАТИСТИКА:")
for key, value in stats.items():
    print(f"{key}: {value:.1f}" if isinstance(value, float) else f"{key}: {value}")
```

```python
# 1. Подсчет распределения по источникам
source_counts = data["ИсточникФайл"].value_counts(normalize=True) * 100

# 2. Format results
report = """
QUESTION DISTRIBUTION BY SOURCES
--------------------------------
"""

for source, percent in source_counts.items():
    report += f"- {source}: {percent:.1f}% of questions\n"

# 3. Distribution analysis
total_sources = len(source_counts)
dominant_source = source_counts.idxmax()
dominant_percent = source_counts.max()
minor_source = source_counts.idxmin()
minor_percent = source_counts.min()

report += f"""
DISTRIBUTION ANALYSIS
--------------------
1. Total source categories: {total_sources}
2. Largest share: {dominant_source} ({dominant_percent:.1f}%)
3. Smallest share: {minor_source} ({minor_percent:.1f}%)
4. Largest/smallest share ratio: {dominant_percent/minor_percent:.1f}:1

KEY OBSERVATIONS:
- {dominant_source} dominates with {dominant_percent:.1f}% of all questions
- {' and '.join(source_counts.index[1:3].tolist())} together account for {source_counts.iloc[1:3].sum():.1f}%
- Distribution shows {'skew' if dominant_percent > 50 else 'balance'} toward {dominant_source.split('.')[0]}
"""

# 4. Output results
print(report)

# 5. Additional statistics (optional)
stats = {
    'Total questions': len(data),
    'Average questions per source': len(data)/total_sources,
    'Median share': source_counts.median()
}

print("\nADDITIONAL STATISTICS:")
for key, value in stats.items():
    print(f"{key}: {value:.1f}" if isinstance(value, float) else f"{key}: {value}")
```
